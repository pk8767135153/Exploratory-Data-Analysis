```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
# sns.get_dataset_names()
```

```python
diamond_data = sns.load_dataset('diamonds')
```

```python
diamond_data.shape
```

### **Basic Plot**

```python
#  Scatter : Visualizing relationships between two continuous variables.

plt.scatter(x = diamond_data["carat"],y = diamond_data["price"],color="skyblue",label="Data Point")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Price_Carat_Relation")
plt.legend()
plt.show()
```

```python
# Bar Plot : To visualize the relationship between a categorical variable and a numerical variable.
plt.bar(diamond_data["cut"].unique(),diamond_data.groupby("cut")["price"].mean(),color = "pink",label = "data.groupby('cut')['price'].mean()")
plt.xlabel("Cuts")
plt.ylabel("Price")
plt.title("Average Prie Of Diamond Cut wise")
plt.legend()
plt.show()
```

```python
diamond_data.head(2)
```

```python
# Histogram : To represent the distribution of a continuous numerical variable. Unlike bar plots.

plt.hist(diamond_data["price"], bins=15,color="skyblue",label="Price",edgecolor="black")
plt.title("Distribution Of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Histogram : To represent the distribution of a continuous numerical variable. Unlike bar plots.

plt.hist([diamond_data["price"],diamond_data["carat"]],bins=6,color=["skyblue","pink"],label=["Price","Carat"],edgecolor="black")
plt.title("Distribution Of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Bar Plot : 1. Compare different groups or categories
#            2. Show changes over time
#            3. Display frequency

plt.boxplot(diamond_data["carat"],vert=False)
plt.title("Box Plot Of Carat")
plt.xlabel("Carat")
plt.show()
```

```python
# Pie Chartn : Calculate Proportions.
cut_count = diamond_data["cut"].value_counts()
plt.pie(cut_count,labels=cut_count.index,autopct='%1.2f%%',startangle=90,explode=[0.05,0.02,0.1,0.2,0.1])
plt.title("Distribution of Cut Quality")
plt.show()
```

### **Advance Plot**

```python
correlation = diamond_data.drop(columns=["price"]).corr()
correlation
```

```python
# Heat Map :
plt.imshow(correlation,cmap="coolwarm",interpolation="none")

# display color bar
plt.colorbar()

plt.xticks(range(len(correlation)),correlation.columns,rotation=90)
plt.yticks(range(len(correlation)),correlation.columns,rotation=90)
plt.title("Heat Map Of Correlation Matrix")
plt.show()
```
